## script used to subset the full set codex data to patient level

start with colon cells

```python
import pandas as pd
import seaborn as sns
```

```python
protein = pd.read_csv("/home/bkzhu/super_mario/hubmap_phase2/codex_cleaned/CL_immune.csv")
protein_stro = pd.read_csv("/home/bkzhu/super_mario/hubmap_phase2/codex_cleaned/CL_stromal.csv")
protein_epi = pd.read_csv("/home/bkzhu/super_mario/hubmap_phase2/codex_cleaned/CL_epi_codex.csv")
protein_epi=protein_epi.rename(columns = {'Cell.Type2':'cluster.term'}) # make sure column name same

print([protein.shape, protein_stro.shape, protein_epi.shape])
```

```python
# combine all the CL cells, and use cells from patient B006
protein_all1 = protein.append(protein_stro)
protein_all = protein_all1.append(protein_epi)
protein_all_B006_CL = protein_all.loc[protein_all["array"] == "B006_CL",:]
```

```python
sns.set(rc={'figure.figsize':(12,12)})
sns.set_theme(style='white')

sns.scatterplot(data=protein_all_B006_CL, x="Xcorr", y="Ycorr", hue="cluster.term", s=0.5)
```
